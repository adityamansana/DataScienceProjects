```python
#extracting information
from bs4 import BeautifulSoup
import requests
url = "https://www.python.org/~guido/"
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html,"html.parser" )  #Create a soup object
```

```python
print(soup.title)   #prints title

print(soup.get_text())  #prints text
```

```python
a_tag_hyperlinks = soup.find_all('a')   #finds all a tags
for link in a_tag_hyperlinks:
    print(link.get("href")) #prints all the URLS
```

```python
img_tag = soup.find_all('img')   #finds all a tags
for image in img_tag:
    print(image.get("src")) #prints all the URLS
```
